Revenue Analysis

```python
# 2. Revenue Analysis:
```

```python
print('Total revenue:',analysis_df['price'].sum().round(2));
print('Average Order value:',analysis_df['price'].mean().round(2));
```

Revenue by Month:

```python
analysis_df['order_purchase_timestamp'] = pd.to_datetime(
    analysis_df['order_purchase_timestamp']) 

monthly_revenue = (
    analysis_df
    .groupby(analysis_df['order_purchase_timestamp'].dt.to_period('M'))['price']
    .sum()
)

monthly_revenue.plot(figsize=(10,5))
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
print(monthly_revenue.idxmax(),monthly_revenue.max())
print(monthly_revenue.idxmin(),monthly_revenue.min())
```

```python
# Insight:
# November 2017 generated the highest monthly revenue (₹1,055,072).
# September 2016 recorded the lowest monthly revenue (₹10.90), 
# primarily due to limited data at the beginning of the dataset.
```

7. Delivery Analysis

Create delivery time:

```python
analysis_df['delivery_time'] = (
    analysis_df['order_delivered_customer_date']
    - analysis_df['order_purchase_timestamp']
).dt.days
```

Average delivery time:

```python
print('Average time:', analysis_df['delivery_time'].mean())
```

```python
analysis_df['delivery_time'].plot(kind='hist', bins=30)

plt.title("Distribution of Delivery Time")
plt.xlabel("Delivery Time (Days)")
plt.ylabel("Number of Orders")
plt.show()
```

```python
# Insights:
# 1. The average delivery time is about 12 days,
#   so most customers receive their orders within two weeks.

# 2. A small number of orders take much longer than usual,
#   which may indicate delivery delays or shipping issues.
```
